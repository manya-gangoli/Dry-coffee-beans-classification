each bean category has many outliers, so first we will have to handle them by feature scalling and for that we have to decide which type of encoding



```python
newdf=df.copy()
newdf['Class'].unique()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
newdf.iloc[:,:-1]=scaler.fit_transform(newdf.iloc[:,:-1])
newdf.head()
#Scaled everything except target class
```

```python
x=newdf.iloc[:,:-1]
y=newdf.iloc[:,-1]
#Split into dependent and independent variables
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.20, random_state=42)
#train test split
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(x_train,y_train)
pred=classifier.predict(x_test)
#to find the accuracy of the model
accuracy=classifier.score(x_test,y_test)
accuracy
```

```python
x1=df.iloc[:,:-1]
y1=df.iloc[:,-1]
#train test split on unstandardized model
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size=0.20, random_state=42)
#train test split
classifier1=SVC(kernel='linear',random_state=0)
classifier1.fit(x1_train,y1_train)
pred1=classifier.predict(x1_test)
#to find the accuracy of the model
accuracy1=classifier1.score(x1_test,y1_test)
accuracy1
```

###after comparing the accuracy of the model before and after standardization we found out that there wasn't much difference in accuracy of the model
